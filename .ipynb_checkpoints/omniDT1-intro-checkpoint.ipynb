## 5. Customization
### Enable Extension
On launch of the Application enable the developer bundle by adding the --dev-bundle or -d flag to the launch command.

```python
!.\repo.bat launch --dev-bundle
```

From the running application select Developer > Extensions

Browse and enable extensions of interest from the Extension Manager.

Enabling the extensions within the Extension Manager UI will allow you to try out the features of the extension in the currently running application.

To permanently add the extension to the application, you will need to add the extension to the .kit file. For example, adding the Layer View extension would require adding omni.kit.widget.layers to the dependencies section of the .kit file.



## 6. Add USD Stage

Change the path to where the sample USD stage is and launch your base editor preloaded with the USD stage.

```python
.\repo.bat launch -- --/app/auto_load_usd="/samples/Rivermark/rivermark.usd"
#--/app/auto_load_usd="\${app}/../samples/stage01.usd"
```

For the remainder of this playbook, you have 3 choices when it comes to the 3D data:
* If you don't have data you can proceed to use the sample scene provided.
  
* If you already have your own USD data to use, you can use that right away by replacing the file path to point to your own file.

* If you have 3D data or want to use your own dataset but it is not in USD format, the __omniDT2__ notebook can guide you on aggregating your dataset and converting it to USD.
